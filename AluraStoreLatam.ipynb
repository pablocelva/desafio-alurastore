### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda_1 = tienda["Precio"].sum()
ingreso_tienda_2 = tienda2["Precio"].sum()
ingreso_tienda_3 = tienda3["Precio"].sum()
ingreso_tienda_4 = tienda4["Precio"].sum()

print(f"El ingreso total de la tienda 1 es: {ingreso_tienda_1}")
print(f"El ingreso total de la tienda 2 es: {ingreso_tienda_2}")
print(f"El ingreso total de la tienda 3 es: {ingreso_tienda_3}")
print(f"El ingreso total de la tienda 4 es: {ingreso_tienda_4}")
```

# 2. Ventas por categoría

```python
categoria_tienda_1 = tienda["Categoría del Producto"].value_counts()
categoria_tienda_2 = tienda2["Categoría del Producto"].value_counts()
categoria_tienda_3 = tienda3["Categoría del Producto"].value_counts()
categoria_tienda_4 = tienda4["Categoría del Producto"].value_counts()
print("Ventas por Categorías Tienda 1: ", categoria_tienda_1)
print("Ventas por Categorías Tienda 2: ", categoria_tienda_2)
print("Ventas por Categorías Tienda 3: ", categoria_tienda_3)
print("Ventas por Categorías Tienda 4: ", categoria_tienda_4)
```

```python
print("Categoría más vendida en Tienda 1: ", categoria_tienda_1.idxmax())
print("Categoría más vendida en Tienda 2: ", categoria_tienda_2.idxmax())
print("Categoría más vendida en Tienda 3: ", categoria_tienda_3.idxmax())
print("Categoría más vendida en Tienda 4: ", categoria_tienda_4.idxmax())
```

# 3. Calificación promedio de la tienda


```python
promedio_calificacion_tienda_1 = tienda["Calificación"].mean()
promedio_calificacion_tienda_2 = tienda2["Calificación"].mean()
promedio_calificacion_tienda_3 = tienda3["Calificación"].mean()
promedio_calificacion_tienda_4 = tienda4["Calificación"].mean()

print("Calificación promedio de la tienda 1: ", promedio_calificacion_tienda_1)
print("Calificación promedio de la tienda 2: ", promedio_calificacion_tienda_2)
print("Calificación promedio de la tienda 3: ", promedio_calificacion_tienda_3)
print("Calificación promedio de la tienda 4: ", promedio_calificacion_tienda_4)
```

# 4. Productos más y menos vendidos

```python
productos_tienda_1 = tienda["Producto"].value_counts()
productos_tienda_2 = tienda2["Producto"].value_counts()
productos_tienda_3 = tienda3["Producto"].value_counts()
productos_tienda_4 = tienda4["Producto"].value_counts()

print("Productos más vendidos en Tienda 1: ", productos_tienda_1.idxmax())
print("Productos menos vendidos en Tienda 1: ", productos_tienda_1.idxmin())

print("Productos más vendidos en Tienda 2: ", productos_tienda_2.idxmax())
print("Productos menos vendidos en Tienda 2: ", productos_tienda_2.idxmin())

print("Productos más vendidos en Tienda 3: ", productos_tienda_3.idxmax())
print("Productos menos vendidos en Tienda 3: ", productos_tienda_3.idxmin())

print("Productos más vendidos en Tienda 4: ", productos_tienda_4.idxmax())
print("Productos menos vendidos en Tienda 4: ", productos_tienda_4.idxmin())
```

```python
print("Productos vendidos en Tienda 1: ", productos_tienda_1)
print("Productos vendidos en Tienda 2: ", productos_tienda_2)
print("Productos vendidos en Tienda 3: ", productos_tienda_3)
print("Productos vendidos en Tienda 4: ", productos_tienda_4)
```

```python
import matplotlib.pyplot as plt

def graficar_mas_vendidos(productos, tienda_nombre):
  productos.head(5).plot(kind="bar", title=f"Productos más vendidos en {tienda_nombre}", figsize=(5, 5))
  plt.xlabel("Productos")
  plt.ylabel("Cantidad vendida")
  plt.xticks(rotation=45)
  plt.tight_layout()
  plt.show()

def graficar_menos_vendidos(productos, tienda_nombre):
  productos.tail(5).plot(kind="bar", title=f"Productos menos vendidos en {tienda_nombre}", figsize=(5, 5))
  plt.xlabel("Productos")
  plt.ylabel("Cantidad vendida")
  plt.xticks(rotation=45)
  plt.tight_layout()
  plt.show()

graficar_mas_vendidos(productos_tienda_1, "Tienda 1")
graficar_menos_vendidos(productos_tienda_1, "Tienda 1")

graficar_mas_vendidos(productos_tienda_2, "Tienda 2")
graficar_menos_vendidos(productos_tienda_2, "Tienda 2")

graficar_mas_vendidos(productos_tienda_3, "Tienda 3")
graficar_menos_vendidos(productos_tienda_3, "Tienda 3")

graficar_mas_vendidos(productos_tienda_4, "Tienda 4")
graficar_menos_vendidos(productos_tienda_4, "Tienda 4")
```
